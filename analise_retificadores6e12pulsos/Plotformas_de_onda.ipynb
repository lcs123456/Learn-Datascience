# Analise da Tensão DC.
## A tensão DC é calculada da seguinte forma:
### Primeiramente determina-se a area que a tensão ocupa e depois a espalha de forma retangular, onde, um lado e definido pelo mesmo tamanho ocupado pela onda no eixo x, o eixo y então representa a tensão média da carga.
#### Desta forma, pode-se falar que a tensão média nesses sistemas podem ser representada pela integral definida, e espalhada:
# $V_{med}=\frac{1}{T}*\int_{\frac{\pi}{6}}^{\frac{2\pi}{6}}V_{12}(\omega t)\partial \omega t$

#### Ao solucionar-mos a integral pode-mos observar que o valor pode ser descrito como:
# $V_{med}=\frac{3\sqrt{3}V_p}{\pi}\cong \sqrt{3}V_p$

# Resposta DC

## Utilizando a formula acima, obtem-se as formas abaixo, nela é possivel observar que que durante 1 ciclo dos 60hz acontece 6 pulsos de tensão, do qual origina o nome deste tipo de retificador: Retificador de 6 pulsos.

```
#@title
plt.plot(lista,DC2)
plt.axis([0,0.016667,0,1700])
plt.title("Resposta DC Ondas Trifásicas Normais (1 Ciclo em 60Hz)", fontsize=20)
plt.xlabel("Tempo(s)", fontsize=15)
plt.ylabel("Tensão(Vcc)", fontsize=15)
plt.show()
```

# Analise com Tempos Maiores.
## Se analisarmos com o tempo de 1s, 60 ciclos, é possivel observar a diferença entre os dois sistemas.
## O ripple maximo é calculado utilizando o maior valor positivo subtraido ao menor valor das ondas e verificado a porcentagem deste valor.
## A formula com a resposta em ripple em porcentagem é:
## $Ripple = \frac{+V_{Pa}-V_{Pb}}{+V_{Pa}}*100 $

```
#@title
plt.plot(lista,DC2)
plt.axis([0,1,0,1700])
plt.title("Resposta DC Ondas Trifásicas Normais Analise em Um Segundo", fontsize=20)
plt.xlabel("Tempo(s)", fontsize=15)
plt.ylabel("Tensão(Vcc)", fontsize=15)
ripple = (max(DC2)-min(DC2))/(max(DC2))
plt.legend(['Ripple Maximo: '+str(round(ripple*100,2))+'%'],fontsize=25, loc=10)
plt.show()
```

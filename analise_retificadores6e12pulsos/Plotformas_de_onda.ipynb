```
from matplotlib import pyplot as plt
import math
import numpy as np
from copy import deepcopy as dc
from scipy.integrate import simps as inte
```

```
lista = np.arange(0,0.016667,0.000001)
```

```
def createlist(angulo = 0):
  a = []
  for x in lista:
    a.append(math.sin(x*366+math.radians(angulo)))
  senoides.append(a.copy())
```

```
senoides = []
createlist(-7.5)
createlist(112.5)
createlist(-127.5)
createlist(22.5)
createlist(142.5)
createlist(-97.5)
```

```
plt.rcParams["figure.figsize"] = (30,10)
plt.plot(lista, senoides[0],lista, senoides[1],lista, senoides[2])
plt.axis([0,0.016667,-1.2,1.2])
plt.title("Ondas Trifasicas Normais")
plt.show()
```

```
plt.rcParams["figure.figsize"] = (30,10)
plt.plot(lista, senoides[0],lista, senoides[1],lista, senoides[2],lista, senoides[3],lista, senoides[4],lista, senoides[5])
plt.axis([0,0.016667,-1.2,1.2])
plt.title("Onda com Duas Saidas Defasadas")
plt.show()
```

```
maxi=[]
mini=[]
for x in range(len(senoides[1])):
  list = [senoides[0][x],senoides[1][x],senoides[2][x],senoides[3][x],senoides[4][x],senoides[5][x]]
  maxi.append(max(list))
  mini.append(min(list))
```

```
maxi2=[]
mini2=[]
for x in range(len(senoides[1])):
  list = [senoides[0][x],senoides[1][x],senoides[2][x]]
  maxi2.append(max(list))
  mini2.append(min(list))
DC1=[]
for x in range(len(maxi)):
  list = [maxi[x],-mini[x]]
  DC1.append(max(list))
DC2=[]
for x in range(len(maxi2)):
  list = [maxi2[x],-mini2[x]]
  DC2.append(max(list))
```

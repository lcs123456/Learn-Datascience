```python
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
```

```python
amazon = pd.read_csv("amazon.csv",encoding='cp1252')
```

```python
amazon.head(25)
```

```python
anos = []
for x in range(len(amazon)):
    if(amazon.iloc[x,3]>0):
        date = amazon.iloc[x,4]
        a=date.split("-")
        for y in range(int(amazon.iloc[x,3])):
            anos.append(a[0])
```

```python
anosN = Counter(anos)
```

```python
anosN
```

```python
from operator import itemgetter
anosN = sorted(anosN.items(), key=itemgetter(0))
```

```python
plt.rcParams['figure.figsize'] = [15, 5]
a=[]
b=[]
for x,y in anosN:
    a.append(x)
    b.append(y)
plt.bar(a,b)
plt.axis([])
plt.show()
```
